```python
%matplotlib inline
import numpy
import matplotlib.pyplot as pp

class MDP:
    def __init__(self, nstates, nactions, P={}, terminals=[0]):
        self.nactions = nactions
        self.nstates = nstates
        self.P = P
        self.terminals = terminals
    def episode(self, policy):
        s = 0
        ep = []
        #rtm1 = 0
        i = 0
        while s not in self.terminals:
            a = policy(s, self)
            dist = self.P[s][a]
            sp = numpy.random.multinomial(1, dist).argmax()
            r = self.P[sp]['r']
            ep.append([s,a,r])
            #rtm1 = r
            s = sp
        ep.append([s,-1,0])
        return ep
        
def firstVisitMC(pi,_, mdp, niterations=100):
    V = numpy.zeros(mdp.nstates)
    nvisits = numpy.zeros(mdp.nstates)
    vs = []
    for it in range(niterations):
        episode = mdp.episode(pi)
        Return = []#numpy.cumsum(episode[:,2][::-1])[::-1]
        G = 0
        # traverse the episode backwards
        for t in range(len(episode)-2,-1,-1): 
            Return.append(G)
            s,a,r = episode[t]
            G = G + r
        Return.append(G)
        Return = numpy.float32(Return)
        episode = numpy.concatenate([episode,Return[::-1][:,None]],axis=1)
        visited = set()
        for s,a,r,G in episode:
            if s in visited: continue
            V[s] += G
            nvisits[s] += 1
        vs.append(V[:-1]/(nvisits[:-1]+1))
    V /= nvisits
    print V
    return V[:-1],numpy.float32(vs)

def everyVisitMC(pi,_, mdp, niterations=100):
    V = numpy.zeros(mdp.nstates)
    nvisits = numpy.zeros(mdp.nstates)
    vs = []
    for it in range(niterations):
        episode = mdp.episode(pi)
        Return = 0
        # traverse the episode backwards
        for t in range(len(episode)-2,-1,-1): 
            s,a,r = episode[t]
            Return = Return + r
            V[s] += Return
            nvisits[s] += 1
        vs.append(V[:-1]/(nvisits[:-1]+1))
    V /= nvisits
    return V[:-1],numpy.float32(vs)

def offPolicyEveryVisitMC(pi, mu, mdp, niterations=100):
    Qsa = numpy.random.random((mdp.nstates, mdp.nactions))
    Csa = numpy.zeros((mdp.nstates, mdp.nactions))
    P = numpy.float32([[pi.prob(s,a,mdp) for a in range(mdp.nactions)] for s in range(mdp.nstates)])
    vs = []
    Ws = []
    for it in range(niterations):
        episode = mdp.episode(mu)
        G = 0
        W = numpy.float64(1)
        # traverse the episode backwards
        for t in range(len(episode)-2,-1,-1): 
            s,a,r = episode[t]
            G = G + r
            Csa[s,a] += W
            Qsa[s,a] += W / Csa[s,a] * (G-Qsa[s,a])
            W = W * pi.prob(s,a,mdp) / mu.prob(s,a,mdp)
            Ws.append(pi.prob(s,a,mdp) / mu.prob(s,a,mdp))
            #print W, s,a, pi.prob(s,a,mdp), mu.prob(s,a,mdp)
            if W == 0:
                break
        
        vs.append((Qsa*P).sum(axis=1)[:-1])
    return (Qsa*P).sum(axis=1)[:-1], numpy.float32(vs)

def offPolicyRecognizers(pi, mu, mdp, niterations=100):
    Qsa = numpy.random.random((mdp.nstates, mdp.nactions))
    Csa = numpy.zeros((mdp.nstates, mdp.nactions))
    RandomRecognizers = rho = numpy.random.random((mdp.nstates, mdp.nactions))
    #rho *= rho > 0.5
    rho /= rho.sum(axis=1)[:,None]
    print rho
    P = numpy.float32([[pi.prob(s,a,mdp) for a in range(mdp.nactions)] for s in range(mdp.nstates)])
    vs = []
    Ws = []
    for it in range(niterations):
        episode = mdp.episode(mu)
        G = 0
        W = numpy.float64(1)
        # traverse the episode backwards
        for t in range(len(episode)-2,-1,-1): 
            s,a,r = episode[t]
            G = G + r
            Csa[s,a] += W
            Qsa[s,a] += W / Csa[s,a] * (G-Qsa[s,a])
            W = W * pi.prob(s,a,mdp) / mu.prob(s,a,mdp)
            Ws.append(pi.prob(s,a,mdp) / mu.prob(s,a,mdp))
            if W == 0:
                break
        
        vs.append((Qsa*P).sum(axis=1)[:-1])
    return (Qsa*P).sum(axis=1)[:-1], numpy.float32(vs)
    


def randompol(s,mdp):
    return numpy.random.randint(0,mdp.nactions)
randompol.prob = lambda s,a,mdp: 1./mdp.nactions


def oddstatesplay0(s,mdp):
    if s%2:
        if numpy.random.random() < 0.7:
            return 0
    return numpy.random.randint(0,mdp.nactions)

def oddstatesplay0_prob(s,a,mdp):
    if s%2:
        if a==0:
            return 0.7 + 1./mdp.nactions*0.3
        return 0.3*1./mdp.nactions
    return 1./mdp.nactions
oddstatesplay0.prob = oddstatesplay0_prob
```

```python

def plot_improvement(pol, mu, mdp, methods,Nmax=3000):
    f,(ax1,ax2) = pp.subplots(1,2,figsize=(16,8))
    V,_=firstVisitMC(pol, mu, mdp,10000)
    ax1.plot([0,100],[V.mean(),V.mean()],label='V_pi')
    ax2.plot([0,Nmax],[V.mean(),V.mean()],label='V_pi')
    for m in methods:
        _,vs=m(pol, mu, mdp,Nmax)
        ax1.plot(vs[:100].mean(axis=1), label=m.func_name)
        ax2.plot(vs.mean(axis=1), label=m.func_name)
    ax1.legend(loc=4)
    ax2.legend(loc=4)
    pp.show()
    
```
